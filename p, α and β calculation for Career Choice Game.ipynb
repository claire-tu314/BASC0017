p, α and β calculation for Career Choice Game

```
import pandas as pd

df = pd.read_excel("compensation_workinghrs_data.xlsx")

# I consider sectors with Average Total Compensation 2023 above $350,000 to be high salary, anything below is low salary.
#Sectors with Weekly Working Hours 2023 above 50 to be weak-fit, below 50 to be strong fit.
salary_threshold = 350000
hours_threshold = 50

# Create classifications
df["Salary_Type"] = df["Average Total Compensation 2023"].apply(lambda x: "H" if x > salary_threshold else "L")

df["Fit_Type"] = df["Weekly Working Hours 2023"].apply(lambda x: "S" if x < hours_threshold else "W")



table = pd.crosstab(df["Fit_Type"], df["Salary_Type"])
print("Contingency table:")
print(table)

# Count the number of sectors of high/low salary and Strong/Weak fit
HS = table.loc["S", "H"] if ("S" in table.index and "H" in table.columns) else 0
S_total = table.loc["S"].sum() if "S" in table.index else 0

HW = table.loc["W", "H"] if ("W" in table.index and "H" in table.columns) else 0
W_total = table.loc["W"].sum() if "W" in table.index else 0

total = len(df)

# Probabilities of fit types
p = S_total / total if total > 0 else 0
one_minus_p = W_total / total if total > 0 else 0

# calculate alpha and beta
alpha = HS / S_total if S_total > 0 else 0
beta = HW / W_total if W_total > 0 else 0

# calculate gamma and mu
HL = table.loc["S", "L"] if ("S" in table.index and "L" in table.columns) else 0
L_total = table["L"].sum() if "L" in table.columns else 0
H_total = table["H"].sum() if "H" in table.columns else 0

mu = HS / H_total if H_total > 0 else 0              # P(S|H)
one_minus_mu = HW / H_total if H_total > 0 else 0    # P(W|H)

gamma = HL / L_total if L_total > 0 else 0           # P(S|L)
one_minus_gamma = (table.loc["W", "L"] if ("W" in table.index and "L" in table.columns) else 0) / L_total if L_total > 0 else 0   # P(W|L)

print("\np =", round(p, 2))
print("1-p =", round(one_minus_p, 2))
print("α =", round(alpha, 2))
print("β  =", round(beta, 2))
print("μ =", round(mu, 2))
print("1-μ =", round(one_minus_mu, 2))
print("γ =", round(gamma, 2))
print("1-γ =", round(one_minus_gamma, 2))
```

```
# Parameters
p = 0.56

# Claire's expected payoffs
def claire_H_accept(mu):
    return 5*mu + 1*(1-mu)

def claire_H_reject(mu):
    return 0*mu + 2*(1-mu)

def claire_L_accept(gamma):
    return 4*gamma + 0*(1-gamma)

def claire_L_reject(gamma):
    return 2*gamma + 3*(1-gamma)

# Threshold checks
def claire_best_after_H(mu):
    if claire_H_accept(mu) > claire_H_reject(mu):
        return "A"
    elif claire_H_accept(mu) < claire_H_reject(mu):
        return "R"
    else:
        return "Indifferent"

def claire_best_after_L(gamma):
    if claire_L_accept(gamma) > claire_L_reject(gamma):
        return "A"
    elif claire_L_accept(gamma) < claire_L_reject(gamma):
        return "R"
    else:
        return "Indifferent"

print("Best response after H when mu = p:", claire_best_after_H(p))
print("Best response after L when gamma = p:", claire_best_after_L(p))

# Pooling on H
mu_pool_H = p
gamma_support_H = 0.0   # any value < 3/5 works

print("\nPooling on H:")
print("mu =", mu_pool_H, "=> Claire after H:", claire_best_after_H(mu_pool_H))
print("gamma =", gamma_support_H, "=> Claire after L:", claire_best_after_L(gamma_support_H))

# Pooling on L
gamma_pool_L = p
mu_support_L = 0.0      # any value < 1/6 works

print("\nPooling on L:")
print("gamma =", gamma_pool_L, "=> Claire after L:", claire_best_after_L(gamma_pool_L))
print("mu =", mu_support_L, "=> Claire after H:", claire_best_after_H(mu_support_L))
```
