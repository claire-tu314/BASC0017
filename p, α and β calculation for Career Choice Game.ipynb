p, α and β calculation for Career Choice Game

```
import pandas as pd

df = pd.read_excel("compensation_workinghrs_data.xlsx")

# Thresholds
salary_threshold = 350000
hours_threshold = 50

# Create classifications
df["Salary_Type"] = df["Average Total Compensation 2023"].apply(
    lambda x: "H" if x > salary_threshold else "L"
)

df["Fit_Type"] = df["Weekly Working Hours 2023"].apply(
    lambda x: "S" if x < hours_threshold else "W"
)

# Contingency table
table = pd.crosstab(df["Fit_Type"], df["Salary_Type"])
print("Contingency table:")
print(table)

# Extract counts safely
HS = table.loc["S", "H"] if ("S" in table.index and "H" in table.columns) else 0
S_total = table.loc["S"].sum() if "S" in table.index else 0

HW = table.loc["W", "H"] if ("W" in table.index and "H" in table.columns) else 0
W_total = table.loc["W"].sum() if "W" in table.index else 0

alpha = HS / S_total if S_total > 0 else 0
beta = HW / W_total if W_total > 0 else 0

print("\nalpha =", round(alpha, 4))
print("beta  =", round(beta, 4))
```
